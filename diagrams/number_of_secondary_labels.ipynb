```python
import pandas as pd
import warnings
warnings.filterwarnings(action='ignore')

train = pd.read_csv('../input/birdclef-2021/train_metadata.csv',)
train['secondary_labels']
```

```python
import re
from collections import defaultdict
d = defaultdict(int)
for s in train['secondary_labels'].tolist():
    res = re.findall(r'\'.*?\'', s)
    for r in res:
        d[r.strip('\'\"')] +=1
```

```python
train['secondary_labels'].values
```

```python
nd = sorted(d.items(), key=lambda x:x[1], reverse=True)
```
